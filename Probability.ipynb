```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import uniform
from scipy.stats import bernoulli
```

```python
# Example: (coin toss probabilities)
tosses = []
for i in range(1,11):
  tosses.append(np.random.randint(0,2))
plt.hist(tosses,bins=2,density=True,range=[0,2])
# area under the curve is the probability, which always equals 1
# density=True shows probabilities, otherwise the charts shows frequencies
```

```python
# Example: (uniform distribution histogram)
a = np.random.choice(np.arange(0,10),1000)
plt.hist(a,bins=np.arange(0,11))
```

```python
# Example: Rolling 2 dice
def roll_two(rolls):
    total = []
    for _ in range(rolls):
        roll_1 = np.random.randint(1,7)
        roll_2 = np.random.randint(1,7)
        total += [(roll_1 + roll_2)]
    return total
rolls = 1000
new_roll = roll_two(rolls)
plt.hist(new_roll,bins=np.arange(2,13,0.5))
plt.show()
```

```python
# Example: Rolling a die (using list comprehension)
ans2a = [x for x in np.random.randint(1,7,1000)]
plt.hist(ans2a,bins=20)
plt.show()
```

```python
# Example: Tossing a coin (using Bernoulli function)
ans1a = 0.5 # probability
ans1b = bernoulli(ans1a) # create a Bernoulli distribution object
ans1c = ans1b.rvs(1000) # generate random samples from the Bernoulli distribution
print(ans1c)
plt.hist(ans1c,bins=20)
plt.show()
```

```python
# Example: (normal distribution histogram)
n_points = 100
np.random.seed(0)
p = np.random.normal(0,1,n_points)
plt.hist(p,density=True)

def myMean(sample):
  N = len(sample)
  total = 0
  for x in sample:
    total = total + x
    return total/N

pMean = myMean(p) # call the function and pass p as the sample
print(f"mean= {pMean}")

def myVar(sample,mean):
  tsample = sample - mean
  var = np.sum(tsample * tsample)
  return var/len(sample)

pVar = myVar(p, pMean)
print(f"variance= {pVar}")

# plot the theortical normal distribution
x = np.linspace(-4,4,n_points) # to create plots for the x-axis
tx = x-pMean
y = np.exp(-(tx-pMean)**2/(2*pVar))/(np.sqrt(2*np.pi*pVar)) # normal distribution formula (not standard)
plt.plot(x,y)
```

```python
#Exmaple: (continous distribution, probability density function (pdf), cumulative distribution function (cdf))
from scipy.stats import norm
x_axis = np.arange(1734, 4357, 1)
plt.plot(x_axis, norm.pdf(x_axis,3026,500))
plt.xlabel("Height")
plt.ylabel("Probability")
plt.show()
H = norm(3026,500)
print(H.cdf(3000)) #computing the the probability density for P(<3000)
```

```python
# Example (Exponential distribution)
from scipy.stats import poisson
ans6a = poisson.rvs(mu=3.54,size=10000)
ax = sns.distplot(ans6a,
                  bins=30,
                  kde = False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
x = np.linspace(0, max(ans6a), 100)


plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')
plt.legend()

# Show the plot
plt.show()

```
